## **Data Preparation**


###**Data Cleaning**

Drop columns ‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô (ID, URL, ‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û ‡∏Ø‡∏•‡∏Ø)

```python
# ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÄ‡∏´‡∏•‡πà‡∏≤‡∏ô‡∏µ‡πâ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡πÇ‡∏¢‡∏ä‡∏ô‡πå‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå ‡πÄ‡∏ä‡πà‡∏ô URL, ‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û, ‡∏£‡∏´‡∏±‡∏™‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô ‡πÅ‡∏•‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥‡∏ã‡πâ‡∏≠‡∏ô

cols_to_drop = [
    'id', 'listing_url', 'scrape_id', 'source', 'name', 'description', 'picture_url',
    'host_url', 'host_thumbnail_url', 'host_picture_url', 'calendar_updated',
    'calendar_last_scraped', 'neighbourhood_group_cleansed', 'license'
]
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
# ‡∏•‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î‡∏Ñ‡πà‡∏≤‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏£‡∏≤‡∏Ñ‡∏≤ ‡∏´‡∏£‡∏∑‡∏≠ ‡πÑ‡∏°‡πà‡∏°‡∏µ room_type ‡∏ñ‡∏∑‡∏≠‡∏ß‡πà‡∏≤‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡πÑ‡∏î‡πâ
df = df.dropna(subset=['price', 'room_type'])
```

‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£ missing values

```python
# ‡∏î‡∏π‡∏à‡∏≥‡∏ô‡∏ß‡∏ô missing
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# Score & Review: ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏£‡∏µ‡∏ß‡∏¥‡∏ß ‡πÉ‡∏´‡πâ‡πÅ‡∏ó‡∏ô‡πÄ‡∏õ‡πá‡∏ô 0

review_cols = [
    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication',
    'review_scores_location', 'review_scores_value', 'reviews_per_month'
]
df[review_cols] = df[review_cols].fillna(0)
```

```python
# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• host: ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå String ‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÑ‡∏õ ‡πÄ‡∏ï‡∏¥‡∏° 'Unknown' ‡∏´‡∏£‡∏∑‡∏≠ 'No description'

df['host_location'] = df['host_location'].fillna('Unknown')
df['host_about'] = df['host_about'].fillna('No description')
df['host_response_time'] = df['host_response_time'].fillna('unknown')
df['neighbourhood'] = df['neighbourhood'].fillna('Unknown')
```

```python
# ‡πÄ‡∏õ‡∏≠‡∏£‡πå‡πÄ‡∏ã‡πá‡∏ô‡∏ï‡πå: ‡πÅ‡∏õ‡∏•‡∏á string ‡πÄ‡∏õ‡πá‡∏ô float ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ï‡∏¥‡∏° median

df['host_response_rate'] = df['host_response_rate'].str.rstrip('%').astype(float)
df['host_response_rate'] = df['host_response_rate'].fillna(df['host_response_rate'].median())

df['host_acceptance_rate'] = df['host_acceptance_rate'].str.rstrip('%').astype(float)
df['host_acceptance_rate'] = df['host_acceptance_rate'].fillna(df['host_acceptance_rate'].median())
```

```python
# ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏´‡πâ‡∏≠‡∏á: ‡πÄ‡∏ï‡∏¥‡∏° median ‡∏Ç‡∏≠‡∏á bathrooms, bedrooms, beds

df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['beds'] = df['beds'].fillna(df['beds'].median())

# drop missing rows
df = df[df['bathrooms_text'].notna()]
```

```python
# Boolean: ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô 1/0 ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î

# ‡πÄ‡∏ï‡∏¥‡∏° 0 ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô superhost ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
df['host_is_superhost'] = df['host_is_superhost'].fillna(0)

# ‡πÄ‡∏ï‡∏¥‡∏° 1 ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ß‡πà‡∏≤ available ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà (‡∏ñ‡∏∑‡∏≠‡∏ß‡πà‡∏≤‡∏°‡∏µ)
df['has_availability'] = df['has_availability'].fillna(1)
```

```python
# ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà: ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô datetime ‡πÅ‡∏•‡πâ‡∏ß‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÄ‡∏™‡∏£‡∏¥‡∏°

# ‡πÅ‡∏õ‡∏•‡∏á‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô datetime ‡πÅ‡∏•‡πâ‡∏ß‡∏™‡∏£‡πâ‡∏≤‡∏á flag ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df['first_review'] = pd.to_datetime(df['first_review'], errors='coerce')

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏°‡∏≤‡πÅ‡∏•‡πâ‡∏ß‡∏à‡∏≤‡∏Å‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î
df['days_since_last_review'] = (pd.to_datetime('today') - df['last_review']).dt.days

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ö‡∏≠‡∏Å‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡πÑ‡∏´‡∏°
df['has_review'] = df['last_review'].notnull().astype(int)
```

```python
# days_since_last_review ‡πÄ‡∏ï‡∏¥‡∏°‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤ -1 ‡πÅ‡∏õ‡∏•‡∏ß‡πà‡∏≤ ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Ñ‡∏¢‡∏°‡∏µ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß

df['days_since_last_review'] = df['days_since_last_review'].fillna(-1)
```

‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•

```python
#  price: ‡∏•‡∏ö $ ‡πÅ‡∏•‡∏∞ , ‡πÅ‡∏•‡πâ‡∏ß‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô float

df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Boolean ('t'/'f'): ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô 1/0

bool_cols = ['host_is_superhost', 'instant_bookable', 'host_identity_verified', 'host_has_profile_pic', 'has_availability']
for col in bool_cols:
    df[col] = df[col].map({'t': 1, 'f': 0})
```

```python
# ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà: ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô datetime

date_cols = ['host_since', 'first_review', 'last_review']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

‡∏•‡∏ö‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥

```python
# ‡∏ï‡∏±‡∏î price ‡∏ó‡∏µ‡πà‡∏ï‡πà‡∏≥‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ (<200) ‡∏´‡∏£‡∏∑‡∏≠‡∏™‡∏π‡∏á‡πÄ‡∏Å‡∏¥‡∏ô‡∏à‡∏£‡∏¥‡∏á (>100,000)
df = df[(df['price'] >= 200) & (df['price'] <= 100000)]

# ‡∏ï‡∏±‡∏î‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà‡πÉ‡∏´‡πâ‡∏û‡∏±‡∏Å‡πÄ‡∏Å‡∏¥‡∏ô 365 ‡∏ß‡∏±‡∏ô ‡∏´‡∏£‡∏∑‡∏≠‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 1 ‡∏Ñ‡∏∑‡∏ô
df = df[(df['minimum_nights'] >= 1) & (df['maximum_nights'] <= 365)]
```

### **Feature Engineering**


‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏ä‡∏¥‡∏á‡∏•‡∏∂‡∏Å

```python
# price_per_person (‡∏£‡∏≤‡∏Ñ‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏ï‡πà‡∏≠‡∏Ñ‡∏ô): ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå affordability

df['price_per_person'] = df['price'] / df['accommodates']
```

```python
# host_days_active: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà host ‡πÄ‡∏õ‡∏¥‡∏î‡∏ö‡∏±‡∏ç‡∏ä‡∏µ

df['host_days_active'] = (pd.to_datetime('today') - df['host_since']).dt.days
```

```python
# rating_category:	‡πÅ‡∏ö‡πà‡∏á‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡πÄ‡∏õ‡πá‡∏ô Excellent / Good / Average / Low

def categorize_rating(score):
    if pd.isnull(score) or score == 0.0:
        return 'No Rating'
    elif score >= 4.8:
        return 'Excellent'
    elif score >= 4.3:
        return 'Good'
    elif score >= 3.5:
        return 'Fair'
    else:
        return 'Poor'

df['rating_category'] = df['review_scores_rating'].apply(categorize_rating)
```

```python
# reviews_per_year: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏ï‡πà‡∏≠‡∏õ‡∏µ

df['reviews_per_year'] = df['number_of_reviews'] / (df['host_days_active'] / 365 + 1)
```

```python
# is_central: ‡∏´‡πâ‡∏≠‡∏á‡∏û‡∏±‡∏Å‡πÉ‡∏ô‡∏¢‡πà‡∏≤‡∏ô‡πÉ‡∏à‡∏Å‡∏•‡∏≤‡∏á‡πÄ‡∏°‡∏∑‡∏≠‡∏á

central_areas = ['Vadhana', 'Khlong Toei', 'Huai Khwang', 'Ratchathewi', 'Sathon']
df['is_central'] = df['neighbourhood_cleansed'].apply(lambda x: 1 if x in central_areas else 0)
```

```python
# room_type_category: ‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏´‡πâ‡∏≠‡∏á‡∏û‡∏±‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°

df['room_type_category'] = df['room_type'].map({
    'Entire home/apt': 'Entire',
    'Private room': 'Private',
    'Shared room': 'Shared',
    'Hotel room': 'Hotel'
})
```

```python
# ‡∏™‡∏¥‡πà‡∏á‡∏≠‡∏≥‡∏ô‡∏ß‡∏¢‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏î‡∏ß‡∏Å‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô
df['has_wifi'] = df['amenities'].str.contains('Wifi').astype(int)
df['has_kitchen'] = df['amenities'].str.contains('Kitchen').astype(int)
df['has_ac'] = df['amenities'].str.contains('Air conditioning').astype(int)
df['has_breakfast'] = df['amenities'].str.contains('Breakfast').astype(int)
df['has_elevator'] = df['amenities'].str.contains('Elevator').astype(int)
df['has_pool'] = df['amenities'].str.contains('Pool').astype(int)
```

```python
# host_experience_level	: ‡πÅ‡∏ö‡πà‡∏á host ‡πÄ‡∏õ‡πá‡∏ô Novice / Intermediate / Expert

def host_level(days):
    if days > 1000:
        return 'Expert'
    elif days > 365:
        return 'Intermediate'
    else:
        return 'Novice'

df['host_experience_level'] = df['host_days_active'].apply(host_level)
```

```python
# price_bucket: ‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏£‡∏≤‡∏Ñ‡∏≤‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ô‡∏≥‡πÄ‡∏™‡∏ô‡∏≠

df['price_bucket'] = pd.cut(df['price'], bins=[0, 1000, 3000, 6000, 10000, np.inf],
                            labels=['Budget', 'Affordable', 'Standard', 'Premium', 'Luxury'])
```

```python
# listing_age_days ‡∏î‡∏π‡∏ß‡πà‡∏≤‡∏´‡πâ‡∏≠‡∏á‡∏û‡∏±‡∏Å‡πÄ‡∏õ‡∏¥‡∏î‡∏°‡∏≤‡∏ô‡∏≤‡∏ô‡πÅ‡∏Ñ‡πà‡πÑ‡∏´‡∏ô‡πÅ‡∏•‡πâ‡∏ß ‡∏°‡∏µ‡∏ú‡∏•‡∏ï‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏ñ‡∏∑‡∏≠‡∏´‡∏£‡∏∑‡∏≠‡∏¢‡∏≠‡∏î‡∏£‡∏µ‡∏ß‡∏¥‡∏ß

df['first_review'] = pd.to_datetime(df['first_review'], errors='coerce')
# Convert today to Timestamp for consistency
today = pd.Timestamp.today()  # Assign today's date to the variable today as a Timestamp
df['listing_age_days'] = (today - df['first_review']).dt.days
# ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ first_review (NaN) ‡πÅ‡∏™‡∏î‡∏á‡∏ß‡πà‡∏≤ listing ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡πÄ‡∏•‡∏¢ ‚Üí ‡∏ï‡∏±‡πâ‡∏á‡πÄ‡∏õ‡πá‡∏ô -1
df['listing_age_days'] = df['listing_age_days'].fillna(-1).astype(int)
```
